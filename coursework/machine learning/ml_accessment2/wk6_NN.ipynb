```python
## sigmoid
def sigmoid(z: torch.Tensor) -> torch.Tensor:
  ### your code here
  z = 1/(1+torch.exp(-z))
  return z

# The derivative of the sigmoid function  ð‘”â€²(ð‘¥ð‘˜)  is defined as  ð‘”â€²(ð‘¥ð‘˜)=ð‘”(ð‘¥ð‘˜)Ã—(1âˆ’ð‘”(ð‘¥ð‘˜))
def sigmoid_derivative(val):
    derivative = val * (1-val)
    return derivative
```

```python
## logistic regression class
class LogisticRegression(nn.Module):
  def __init__(self, num_features):
    super().__init__()
    self.weight = nn.Parameter(torch.rand(1, num_features), requires_grad=False)

  def forward(self, x):
    y = 0
    ### your code here
    x = x.view(1, -1)
    logistic_x = torch.mm(x, self.weight.t())
    y = sigmoid(logistic_x)
    return y
```

```python
class NeuralNetwork(nn.Module):
  def __init__(self, num_inputs, num_neurons, num_outputs):
    super().__init__()
    self.hidden_layer = nn.ModuleList([LogisticRegression(num_inputs) for i in range(num_neurons)])
    self.output_layer = nn.ModuleList([LogisticRegression(num_neurons+1) for i in range(num_outputs)]) # +1 for the bias

  def forward(self, x: torch.tensor) -> torch.tensor:
    y_hidden = torch.ones(len(self.hidden_layer) + 1) # +1 for the bias
    ### your code here

    # Step 1: Forward pass through the hidden layer
    for(i, hiddenLayer) in enumerate(self.hidden_layer):
        y_hidden[i] = hiddenLayer(x)

    y_out = torch.ones(len(self.output_layer))
    # Step 2: Forward pass through the output layer
    for(i, outputLayer) in enumerate(self.output_layer):
        y_out[i] = outputLayer(y_hidden)

    y=y_out
    return y

  def backward_pass(self, x: torch.tensor, y: torch.tensor, lr: float) -> None:
    ### your code here

    y_pred = self.forward(x)

    # Step 1: For each output, k, calculate the partial derivative ð›¿ð‘˜=(ð‘¦ð‘˜âˆ’ð‘¡ð‘˜)Ã—ð‘”â€²(ð‘¥ð‘˜)
    output_deltas = (y_pred - y) * sigmoid_derivative(y_pred)

    # Step 2: propagate error to the hidden neurons ð›¿ð‘—=ð‘”â€²(ð‘¥ð‘—)âˆ‘ð‘˜ð‘¤ð‘—ð‘˜ð›¿ð‘˜
    hidden_deltas = []
    for j, hidden_layer in enumerate(self.hidden_layer):
       error = 0
       for k, output_layer in enumerate(self.output_layer):
          error += output_layer.weight[0,j] * output_deltas[k]
       hidden_deltas.append(sigmoid_derivative(hidden_layer(x)) * error)

    # Step 3: update the output weights, ð‘¤ð‘—ð‘˜=ð‘¤ð‘—ð‘˜âˆ’ðœ‚ð›¿ð‘˜ð›¼ð‘—
    for k, output_layer in enumerate(self.output_layer):
       weight = output_layer.weight
       for j, hidden_layer in enumerate(self.hidden_layer):
          weight[0,j] = weight[0,j] - lr * output_deltas[k] * hidden_layer(x)
       weight[0,-1] = weight[0,j] - lr * output_deltas[k]

    # Step 4: update the hidden weights ð‘¤ð‘–ð‘—=ð‘¤ð‘–ð‘—âˆ’ðœ‚ð›¿ð‘—ð‘¥ð‘–
    for j, hidden_layer in enumerate(self.hidden_layer):
       weight = hidden_layer.weight
       for i in range(len(x)):
          weight[0, i] = weight[0, i] - lr * hidden_deltas[j] * x[i]  # Perform element-wise multiplication
```

In the `NeuralNetwork` class above, fill in the `backward_pass()` method.

The implementation should support outputs of any size. To implement the backward pass, follow the steps below:



*   **Step 1:** For each output, $k$, calculate the partial derivative: $\delta_{k} = (y_{k}-t_{k}) \times g'(x_{k})$ \\
where $y_{k}$ is the response of the output neuron and $t_{k}$ is the desired output (target). The derivative of the sigmoid function $g'(x_{k})$ is defined as $g'(x_{k}) = g(x_{k}) Ã— (1 - g(x_{k}))$ [1 marks]
*  **Step 2:** We now need to probagate this error to the hidden neurons. To accomplish this remember that: $$
 \delta_{j} = g^\prime(x_{j}) \sum_{k} w_{jk} \delta_{k}
$$  where $\delta_{j}$ is the error on the j-th hidden neuron,$x_{j}$ is the value of the hidden neuron (before it has been passed through the sigmoid function), $g^\prime$ is the derivative of the sigmoid function, $\delta_{k}$ is the error from the output neuron calculated in step 1, and $w_{jk}$ is the weight from the hidden neuron $j$ to the output neuron $k$. [1 marks]

* **Step 3:** We now need to update the output weights, i.e. the connections from the hidden neurons to the output neurons. This is accomplished using the formula: $$
 w_{jk} = w_{jk} - \eta \delta_{k} \alpha_{j}
$$ where $w_{jk}$ is the weight connecting the j-th hidden neuron to the k-th output neuron. $\alpha_{j}$ is the activity of the j-th hidden neuron (after it has been transformed by the sigmoid function), $\delta_{k}$ is the error from the output neuron stored in output\_deltas and $\eta$ is the learning rate [1 mark]

* **Step 4:**  Finally we need to update the hidden weights, i.e. the connections from the hidden neurons to the inputs. Here, again we use this equation $$
 w_{ij} = w_{ij} - \eta \delta_{j} x_{i}
$$ where $w_{ij}$ is the weight connecting the i-th input to the j-th hidden neuron. $x_{i}$ is the i-th input, $\delta_{j}$ is the backpropagated error (i.e., hidden deltas) from the j-th hidden neuron and $\eta$ is the learning rate. [1 mark]

```python
def bce(y_true: torch.Tensor, y_pred: torch.Tensor) -> torch.Tensor:
  ### your code here
  res = torch.mean(-y_true*torch.log(y_pred)-(1-y_true)*torch.log(1-y_pred))
  return res

x1 = torch.tensor([0., 0., 1., 1.])
x2 = torch.tensor([0., 1., 0., 1.])
bias = torch.ones_like(x1)
x = torch.stack([x1, x2, bias]).permute(1, 0)
print(x.shape)
y = torch.tensor([0., 1., 1., 0.])

lr = 1
iterations = 10000
model = NeuralNetwork(num_inputs=3, num_neurons=2, num_outputs=1)

errors = list()
for ep in range(iterations):
  error = 0
  for i in range(x.shape[0]):
    y_pred = model(x[i])
    error += bce(y[i], y_pred)
    model.backward_pass(x[i], y[i], lr)
  errors.append(error/x.shape[0])
plt.plot(list(range(iterations)), [error.detach().numpy() for error in errors])
plt.show()

y_pred = torch.zeros_like(y)
for i in range(x.shape[1]):
  y_pred[i] = model(x[i])
print(y, y_pred)

print('Minimum cost: {}'.format(min(errors)))
```

```python
iris = datasets.load_iris()
print(iris.DESCR)
```

Split the data to train and test sets (make sure the same random seed is used as previously) and normalize using the method from week 4. We will use all attributes in this lab. [2 marks]

```python
### your code here
X_train, X_test, y_train, y_test = model_selection.train_test_split(
    iris.data,
    iris.target,
    test_size=0.2,
    random_state=42
    )
x_train = torch.from_numpy(X_train).float()
x_test = torch.from_numpy(X_test).float()

y_train = torch.from_numpy(y_train).int()
y_train = y_train.reshape(-1, 1)

y_test = torch.from_numpy(y_test).int()
y_test = y_test.reshape(-1, 1)


def norm_set(x: torch.Tensor, mu: torch.Tensor, sigma: torch.Tensor) -> torch.tensor:
    norm_val=(x-mu)/sigma
    return norm_val

mu_val = torch.mean(x_train,dim=0)
std = torch.std(x_train,dim=0)

x_train=norm_set(x_train, mu_val, std)
x_test=norm_set(x_test, mu_val, std)
y_train = torch.tensor(y_train.flatten(), dtype=torch.long)
y_test = torch.tensor(y_test.flatten(), dtype=torch.long)
```

Using pytorch built in methods (and using the training loop from week 4 as guideline for the train loop), build an MLP with one hidden layer. Train the network multiple times for the following number of hidden neurons `{1, 2, 4, 8, 16, 32}`.

<u><font color="maroon"> **Q3.** Explain the performance of the different networks on the training and test sets. How does it compare to the logistic regression example? Make sure that the data you are refering to is clearly presented and appropriately labeled in the report. [8 marks]</font></u>

```python
class MLP(nn.Module):
    def __init__(self, num_inputs, num_neurons, num_outputs):
        super(MLP, self).__init__()
        self.hiddenLayer = nn.Linear(num_inputs, num_neurons)
        self.sigmoid = nn.Sigmoid()
        self.outputLayer = nn.Linear(num_neurons, num_outputs)

    def forward(self, x):
        y = self.hiddenLayer(x)
        y = self.sigmoid(y)
        y = self.outputLayer(y)
        return y
```

```python
criterion = nn.CrossEntropyLoss()

def train(model, x, y, optimizer, alpha, iterations):
    loss_list = []

    for i in range(iterations):
        optimizer.zero_grad()
        pred = model(x)
        cost = criterion(pred, y)
        cost.backward()
        loss_list.append(cost.item())
        optimizer.step()

    return loss_list
```

```python
unique_values = torch.unique(y_test)
num_y_unique_values = len(unique_values)
input_size = x_train.shape[1]
output_size = num_y_unique_values

learning_rate=0.01
hidden_neurons_list = [1, 2, 4, 8, 16, 32]

for hidden_neurons in hidden_neurons_list:
    model = MLP(input_size, hidden_neurons, output_size)
    optimizer = optim.SGD(model.parameters(), lr=learning_rate)
    loss_list=train(model, x_train, y_train, optimizer, learning_rate, iterations)

    with torch.no_grad():
        model.eval()
        y_output = model(x_test)
        predicted = torch.max(y_output.data, 1)[1]
        accuracy = (predicted == y_test).sum().item()/y_test.size(0)
        print(f'Hidden Neurons: {hidden_neurons}, Accuracy on Test Data: {100 * accuracy:.2f}%')

    plt.plot(range(iterations), loss_list, label = f"Hiden Neuron {hidden_neurons}")
    plt.ylabel('cost')
    plt.xlabel('iterations')
plt.legend()
plt.show()
```

```python
for hidden_neurons in hidden_neurons_list:
    model = MLP(input_size, hidden_neurons, output_size)
    optimizer = optim.SGD(model.parameters(), lr=learning_rate)
    train_loss_list=train(model, x_train, y_train, optimizer, learning_rate, iterations)
    test_loss_list=train(model, x_test, y_test, optimizer, learning_rate, iterations)

    plt.plot(range(iterations), train_loss_list, label = f"train cost {hidden_neurons}")
    plt.plot(range(iterations), test_loss_list, label = f"test cost {hidden_neurons}")
    plt.title(f"hidden neurons {hidden_neurons}")
    plt.ylabel('cost')
    plt.xlabel('iterations')
    plt.legend()
    plt.show()
```

```python
class LR(nn.Module):
  def __init__(self,input_dim,output_dim):
    super(LR, self).__init__()
    self.layer = nn.Linear(input_dim,output_dim)
    self.sigmoid = nn.Sigmoid()

  def forward(self,x):
    output = self.sigmoid(self.layer(x))
    return output

input_size = x_train.shape[1]
output_size = len(torch.unique(y_train))
model = LR(input_size, output_size)
optimizer = optim.SGD(model.parameters(), lr=learning_rate)
loss_list=train(model, x_train, y_train, optimizer, learning_rate, iterations)

with torch.no_grad():
    model.eval()
    y_output = model(x_test)
    predicted = torch.max(y_output.data, 1)[1]
    accuracy = (predicted == y_test).sum().item()/y_test.size(0)
    print(f'Accuracy on Test Data: {100 * accuracy:.2f}%,', ' Minimum cost: {}'.format(min(loss_list)))
```
